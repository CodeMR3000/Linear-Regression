```python
def slope(X, Y, m, j):
  M = X.shape[0]
  s = 0
  for i in range(M):
    s += (-2/M)*(Y[i,0] - (m*X[i]).sum())*X[i, j]
  return s
```

```python
def gd(X, Y, learning_rate, num_iterations):
  M = X.shape[0]
  N = X.shape[1]
  m = np.zeros(N)

  for i in range(num_iterations):
    for j in range(N):
      m[j] = m[j] - learning_rate*slope(X, Y, m, j)
  return m
```

```python
def gradient_descent(X, Y, learning_rate, num_iterations):
  M = X.shape[0]
  N = X.shape[1]
  m = np.zeros(N)
  for i in range(num_iterations):
    for j in range(M):
      for k in range(N):
        m[k] = m[k] - (learning_rate * ( (-2/M) * ( Y[j,0] -  (m*X[j]).sum() ) * X[j, k] ))
    # print(m[0])
  return m
```

```python
def predict(X_test, m):
  M = X_test.shape[0]
  Y = np.zeros((M, 1))

  for i in range(M):
    Y[i,0] = round((m*X_test[i]).sum(), 1)
  return Y
```

```python
def run():
  X = df.iloc[:, 0:13].values
  Y = df.iloc[:, 13:].values
  X = np.append(X, [[1] for i in range(379)], axis = 1)
  print(X.shape)
  num_iterations = 1000
  learning_rate = 0.01
  m = gradient_descent(X,Y,learning_rate, num_iterations)
  print(m)
  df2 = pd.read_csv('/content/drive/MyDrive/datasets/test_x_boston.csv', header = None)
  X_test = df2.values
  X_test = np.append(X_test, [[1] for i in range(127)], axis = 1)
  
  Y = predict(X_test, m)
  print(Y)
  np.savetxt("pred.csv", Y, delimiter=",")

```
